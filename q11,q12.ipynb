Create a function for all the tables

```python
def cleanstate(state):
    if "&" in state:
        return state.replace("&","and")
    if state=="Odisha":
        state="Orissa"
        return state
    if state=="Puducherry":
        state="Pondicherry"
        return state
    if state=="Nct Of Delhi":
        state="Delhi"
        return state
    else:
        return state
```

```python
census['State/UT']=census['State/UT'].apply(cleanstate)
```

```python
housing['State Name']=housing['State Name'].apply(cleanstate)
```

```python
hospitals['State/UT']=hospitals['State/UT'].apply(cleanstate)
```
