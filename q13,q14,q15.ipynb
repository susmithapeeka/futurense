```python
government_hospitals['Last_Updated']=pd.to_datetime(government_hospitals['Last_Updated'],format="%d.%m.%Y")
```

```python
#function in previous file 

def cleanstate(state):
    if "&" in state:
        return state.replace("&","and")
    if state=="Odisha":
        state="Orissa"
        return state
    if state=="Puducherry":
        state="Pondicherry"
        return state
    if state=="Nct Of Delhi":
        state="Delhi"
        return state
    else:
        return state.title()
    
```

```python
government_hospitals['State/UT']=government_hospitals['State/UT'].str.replace("*","")
```

```python
government_hospitals['State/UT']=government_hospitals['State/UT'].apply(cleanstate)
```
