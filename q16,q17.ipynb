Problem Statement 16: (Gap in number of beds)

Visually represent the difference between the expected number of hospital beds and the available number of hospital beds in each State/UT, as well as at the national level (if any).


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
hospital=pd.read_csv('clean_Data/otherclean/hospital_clean.csv',index_col='Unnamed: 0')
```

```python
hospital.head()
```

```python
hospital['Required']=(hospital['Population']*3)/1000
```

```python
hospital
```

```python
onlystates=hospital.iloc[0:36]
```

```python
onlystates.to_csv('clean_Data/otherclean/statesCleanq16.csv')
```

```python
states=onlystates['State/UT'].tolist()
```

```python
states
```

```python
presentBeds=onlystates['HospitalBeds'].tolist()
```

```python
RequiredBeds=onlystates['Required'].tolist()
```

```python
pb=hospital['HospitalBeds'].tolist()
```

```python
rb=hospital['Required'].tolist()
```

https://www.geeksforgeeks.org/plotting-multiple-bar-charts-using-matplotlib-in-python/

```python
X_axis = np.arange(len(states))
plt.figure(figsize=(20,15))
plt.bar(X_axis - 0.2, presentBeds, 0.4, label = 'available',color='#FFE0AC')
plt.bar(X_axis + 0.2, RequiredBeds, 0.4, label = 'expected',color='#FFACB7')
plt.xticks(X_axis, states,rotation=90)
plt.yticks(np.arange(round(min(RequiredBeds)),round(max(RequiredBeds)),50000))
plt.xlabel("States")
plt.ylabel("Beds")
plt.title("Required vs available beds in India")
plt.legend(loc=0)
plt.savefig('expectedbeds_chart.png')
plt.show()
```

In a hypothetical situation if the government wishes to fill the gap between the available facility and the WHO standards all by itself how many government hospitals may be required in each state or union territory as well as at the national level?
Find the average number of beds in a government hospital. Divide it by the gap in the number of beds to reach the standards in the region, and round the number to the nearest integer to find the value.
Represent the findings visually.

```python
government_hospitalsavg=hospital['HospitalBeds'].tolist()
```

```python
government_hospitalsavg=sum(government_hospitalsavg)/len(government_hospitalsavg)
```

```python
government_hospitalsavg
```

```python
hospital['gap']=hospital['Required']-hospital['HospitalBeds']
```

```python
hospital=hospital[0:36]
```

```python
hospital.info()
```

```python
hospital['Hospitals']=hospital['Hospitals'].astype('int64')
```

```python
hospital[['HospitalBeds','Hospitals']]
```

```python
hospital['bedsperhospital']=hospital['HospitalBeds']/hospital['Hospitals']
```

```python
hospital.head()
```

```python
hospital.rename(columns={'Per1000':'bedsper10k','bedsperhospital':'avg beds'},inplace=True)
```

```python
hospital['new hospitals']=round(hospital['gap']/hospital['avg beds'])
```

```python
 hospital.head()
```

```python
hospital.to_csv('clean_Data/otherclean/hospitalq17.csv')
```

```python
x = hospital['State/UT'].tolist()
y1 = hospital['Hospitals'].tolist()
y2 = hospital['new hospitals'].tolist()
plt.title('Gap or average of beds')
plt.bar(x, y2, bottom=y1,color='#FFACB7')
plt.bar(x, y1, color='#FFE0AC')
plt.legend(['Present','Required/Gap'],loc=0)
plt.xticks(rotation=90)
plt.show()
```
