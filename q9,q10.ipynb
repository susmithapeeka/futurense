After the updates, the “Households_Rural” and “Households_Urban” data are redundant in both housing and census data.
Compare the two data and see if there is a major difference between the values in the two files.
A difference of more than 10% can be considered a major difference. Report the names of the districts where a major difference is found in the data.

```python
import pandas as pd
import numpy as np
```

```python
housing=pd.read_csv('clean_Data\Clean_Data\housing_cleanwithStates.csv')
```

```python
housing.columns
```

```python
census=pd.read_csv('clean_Data\census_clean.csv')
```

```python
census_sorted=census.groupby(by=['State/UT','District']).sum()
census_sorted
```

```python
housing_sorted = housing.groupby(by=['State/UT','District']).sum()
housing_sorted
```

```python
household_rural_diff=((abs(housing_sorted['Households_Rural']-census_sorted['Households_Rural']))/census_sorted['Households_Rural'])*100
ind=household_rural_diff[household_rural_diff>10].index
ind
```

```python
household_urban_diff=((abs(housing_sorted['Households_Urban']-census_sorted['Households_Urban']))/census_sorted['Households_Urban'])*100
ind=household_urban_diff[household_urban_diff>10].index
ind
```
